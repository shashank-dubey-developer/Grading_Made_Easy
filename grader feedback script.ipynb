```python
from docx import Document
import os
import xlwt
from xlwt import Workbook
import re
import sys 

wb = Workbook()
sheet1 = wb.add_sheet('Sheet 1')
for i in range(14):
    sheet1.write(i,0,("Group Number "+str(i+1)))


# Load the first table from your document. In your example file,
# there is only one table, so I just grab the first one.
directoy = './Docs'
excelCol=1
group_number = []
final_score=[]
for filename in os.listdir(directoy):
    print(filename)
    document = Document("./Docs/"+filename)
    
    
    for par in document.paragraphs:
        if 'Group Name:' in par.text:
            name = par.text.split("Group Name:")[1]
            
            name = re.sub("[^0-9]", "", name).strip()
            if name=='' or name==None or name==' ':
                break
            group_number.append(int(name))
            # print(name)
            # group_number.append(par.text.split(" Group Name:")[1].strip().split(" ")[1])
    
    for i in range(13):
    
        table = document.tables[i]

# Data will be a list of rows represented as dictionaries
# containing each row's data.
        data = []

        keys = None
        for i, row in enumerate(table.columns):
            text = (cell.text for cell in row.cells)
            
    # Establish the mapping based on the first row
    # headers; these will become the keys of our dictionary
            if i == 0:
                keys = tuple(text)
                continue

    # Construct a dictionary for this row, mapping
    # keys to values for this row
            row_data = dict(zip(keys, text))
            data.append(row_data)
        sValues = []
        values=[]
    
        for key,value in data[4].items():
            sValues.append(value)
        
        for i in range(1, len(sValues)):
            if sValues[i]=='':
                sValues[i]=0
            sValues[i] = re.sub("[^0-9]", "", str(sValues[i]))
            values.append(int(sValues[i]))

        final_score.append(sum(values))

    group_score = {}
    
    for i in range(len(group_number)):
        group_score[group_number[i]]=final_score[i]
    print(group_score)
# Workbook is created
    


# add_sheet is used to create sheet.

    for key in group_score:
        sheet1.write(int(key)-1,excelCol,group_score[key])  
    
    wb.save('xlwt example.xls')
    excelCol+=1
```
